```python
import pandas as pd
```

```python
customer_list = pd.read_excel('KPMG_VI.xlsx', sheet_name='CustomerDemographic')
```

```python
df_transactions = pd.read_excel('KPMG_VI.xlsx', sheet_name='Transactions')
```

```python
# Removed transactions done by customers that are not in the registered in the master table
df_transactions = df_transactions.merge(customer_list.customer_id, on='customer_id')
# Order by transaction id
#df_transactions = df_transactions.sort_values('transaction_id')
# Reset indexes
df_transactions.reset_index(drop=True, inplace=True)
```

```python
df_transactions
```

```python
customer_list = customer_list.merge(df_transactions[['customer_id', 'recency']].groupby('customer_id', 
                                                                                       as_index = False).min(), how='outer')

customer_list = customer_list.merge(df_transactions[['customer_id', 'profit']].groupby('customer_id', 
                                                                                       as_index = False).count(), how='outer')
customer_list.columns=list(customer_list.columns[:-1])+['frequency']

customer_list = customer_list.merge(df_transactions[['customer_id', 'profit']].groupby('customer_id', 
                                                                                       as_index = False).mean(), how='outer')
customer_list.columns=list(customer_list.columns[:-1])+['monetary']
```

```python
customer_list
```

```python
# Drop since they are just 87 out of 4000 (~2%) samples
customer_list.drop(customer_list.index[customer_list.recency.isna()], inplace = True)
# Reset indexes
customer_list.reset_index(drop=True, inplace=True)
```

```python
customer_list
```

```python
customer_list[['recency', 'frequency', 'monetary']].describe()
```

```python
customer_list['R_score'] = pd.qcut(customer_list.recency, 4, labels=[4, 3, 2, 1])
customer_list['F_score'] = pd.qcut(customer_list.frequency, 4, labels=[1, 2, 3, 4])
customer_list['M_score'] = pd.qcut(customer_list.monetary, 4, labels=[1, 2, 3, 4])
```

```python
customer_list.drop(columns=['monetary', 'frequency', 'recency'], inplace=True)
```

```python
for cat in ['R_score', 'F_score', 'M_score']:
    customer_list[cat] = pd.to_numeric(customer_list[cat])
```

```python
customer_list['RFM']=100*customer_list.R_score+10*customer_list.F_score+customer_list.M_score
```

```python
customer_list.drop(columns=['R_score', 'F_score', 'M_score'], inplace=True)
customer_list.RFM.describe()
```

```python
customer_list['Status'] = pd.qcut(customer_list.RFM, 4, labels=['Bronze', 'Silver', 'Gold', 'Platinum'])
```

```python
customer_list.drop(columns='RFM', inplace=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(customer_list.drop(columns=['customer_id', 'Status']), customer_list.Status,
                                                    test_size=0.2, random_state=42)
```

```python
clf = RandomForestClassifier(n_estimators=1000)
```

```python
X_train
```

```python
clf.fit(X_train, y_train)
```

```python
clf.feature_importances_
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
cross_val_score(clf, X_train, y_train, cv = 3, scoring='accuracy')
```

```python
confusion_matrix(y_test, clf.predict(X_test))
```
